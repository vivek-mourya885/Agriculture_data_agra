# Plotting the Patterns for Different markets

```python
Samsabad_price = book[book["Market Name"] == 'Samsabad']
y = Samsabad_price['Min Price (Rs./Quintal)']
x = Samsabad_price['Price Date']

plt.plot(x, y)
plt.title('Variation of Min Price in year 2020 for Samsabad Market')
plt.ylabel('Min Price---->')
plt.xlabel('Price Date ----->')
plt.show()
```
